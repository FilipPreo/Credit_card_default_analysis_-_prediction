```python
df = df.loc[~df.artist_genre.isna()]
```

```python
def classify_genres(df, col):
    uniques=[]
    for row in df[col]:  
        for j in row:
            new_term=''
            if j==' ':
                if new_term in uniques:
                    df[new_term] = 1
                else:
                    uniques.append(new_term)
                    df[new_term] = 1
                continue
            else: 
                new_term+=j
                
    return df
```

```python
classify_genres(df, 'artist_genre')
```
